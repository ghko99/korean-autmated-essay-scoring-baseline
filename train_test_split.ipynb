```python
import pandas as pd

dataset = pd.read_csv('./nikl/dataset.csv')

dataset
```

```python
Q1 = dataset[dataset['prompt_num'] == 'Q1']
Q2 = dataset[dataset['prompt_num'] == 'Q2']
Q3 = dataset[dataset['prompt_num'] == 'Q3']
Q4 = dataset[dataset['prompt_num'] == 'Q4']
Q5 = dataset[dataset['prompt_num'] == 'Q5']
Q6 = dataset[dataset['prompt_num'] == 'Q6']
Q7 = dataset[dataset['prompt_num'] == 'Q7']
Q8 = dataset[dataset['prompt_num'] == 'Q8']
Q9 = dataset[dataset['prompt_num'] == 'Q9']

Q1
```

```python
from sklearn.model_selection import train_test_split

Q1_train, Q1_test = train_test_split(Q1, test_size=0.1, random_state=42)
Q2_train, Q2_test = train_test_split(Q2, test_size=0.1, random_state=42)
Q3_train, Q3_test = train_test_split(Q3, test_size=0.1, random_state=42)
Q4_train, Q4_test = train_test_split(Q4, test_size=0.1, random_state=42)
Q5_train, Q5_test = train_test_split(Q5, test_size=0.1, random_state=42)
Q6_train, Q6_test = train_test_split(Q6, test_size=0.1, random_state=42)
Q7_train, Q7_test = train_test_split(Q7, test_size=0.1, random_state=42)
Q8_train, Q8_test = train_test_split(Q8, test_size=0.1, random_state=42)
Q9_train, Q9_test = train_test_split(Q9, test_size=0.1, random_state=42)
```

```python
train = pd.concat([Q1_train, Q2_train, Q3_train, Q4_train, Q5_train, Q6_train, Q7_train, Q8_train, Q9_train])
test = pd.concat([Q1_test, Q2_test, Q3_test, Q4_test, Q5_test, Q6_test, Q7_test, Q8_test, Q9_test]) 


train.to_csv('./nikl/train.csv', index=False, encoding='utf-8-sig')
test.to_csv('./nikl/test.csv', index=False, encoding='utf-8-sig')
```
